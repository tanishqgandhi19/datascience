# Excersise on various Visualization Techniques

The iris dataset contains measurements for 150 iris flowers from three different species.


The three classes in the Iris dataset:
Iris-setosa (n=50)
Iris-versicolor (n=50)
Iris-virginica (n=50)
The four features of the Iris dataset:
sepal length in cm
sepal width in cm
petal length in cm
petal width in cm


Get the data
*Use seaborn to get the iris data by using: iris = sns.load_dataset('iris') *



```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species',hue_order=['virginica','versicolor','setosa'])
```

Create a kde plot of sepal_length versus sepal width for setosa species of flower.

```python
sns.kdeplot(iris[iris.species == "setosa"]["sepal_width"], iris[iris.species == "setosa"]["sepal_length"], 
            shade = True, cmap = "plasma", shade_lowest = False)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="box", data=tips);
```

```python
sns.jointplot('total_bill','tip',data=tips,kind='hex')
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg')
```

```python
x = np.linspace(0,10,100)
y = x
plt.plot(x, y, '-b', label='linear')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.linspace(1.0, 2.0, 1000)
y = 10 * x
plt.plot(x, y, color = "steelblue")
s = np.linspace(2.0, 4.0, 1000)
t = (5 * s) + 10
plt.plot(s, t, color = "steelblue")

plt.plot(4, 27, marker = "^", color = "green")
plt.plot(2.25, 23, marker = "^", color = "green")
plt.plot(1.35, 12, marker = "^", color = "green")

plt.xlim(0.5, 4.5)
plt.ylim(0, 35)
```

```python
x = np.linspace(0,10, 1000)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), linestyle = "dashed")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,5))
x = np.linspace(0.1, 10, 1000)

ax1.plot(x, 0 * x)
ax1.plot(x, np.exp(x))

ax2.plot(x, np.log(x))
plt.yscale("log")
ax2.plot(x, np.exp(x))
```

```python
x = np.linspace(-0.75, 1, 1000)
y = x ** 3

f = plt.figure()
axis = f.add_subplot(1, 1, 1)

axis.spines["left"].set_position("center")
axis.spines["bottom"].set_position("zero")
axis.spines["right"].set_color("none")
axis.spines["top"].set_color("none")

axis.xaxis.set_ticks_position("bottom")
axis.yaxis.set_ticks_position("left")
plt.plot(x,y)
```
